# One-dimensional distribution

The coupling spline approach does not work one-dimensional distributions. However, one learn a one-dimensional distribution by augmenting it with an arbitrary random dummy variable. A good choice is to draw the second variable from the latent distribution, so that the bijector just has to learn the identity mapping for the augmented variable.

We also use this example to learn a distribution with a fat tail.

```python
from neural_flow import Flow, train
from neural_flow.bijectors import rolling_spline_coupling
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

rng = np.random.default_rng(1)

x = rng.pareto(5, size=1000)

fig, ax = plt.subplots(1, 2, figsize=(10, 4))
ax[0].hist(x, bins=100)
ax[1].hist(x, bins=100)
plt.semilogy();
```

To transform 1D distribution, we augment the second dimension with a toy variable. We draw it from the uniform distribution.

```python
flow = Flow(rolling_spline_coupling(2))

X = np.column_stack((x, rng.uniform(size=len(x))))

X_train, X_test = train_test_split(X)

vars, best_epoch, loss_train, loss_test = train(flow, X_train, X_test)

plt.plot(loss_train, label="train")
plt.plot(loss_test, label="test")
plt.legend();
```

```python
xp = flow.apply(vars, 10000, method="sample")[..., 0]
fig, ax = plt.subplots(1, 2, figsize=(10, 4))
ax[0].hist(x, bins=100, density=True, alpha=0.5, label="original")
ax[0].hist(xp, bins=100, density=True, alpha=0.5, label="predicted")
ax[1].hist(x, bins=100, density=True, alpha=0.5, label="original")
ax[1].hist(xp, bins=100, density=True, alpha=0.5, label="predicted")
plt.semilogy()
plt.legend();
```
