```python
from neural_flow.distributions import Normal, Uniform, TruncatedNormal, Beta
from matplotlib import pyplot as plt
import numpy as np
```

```python
x = np.linspace(-6, 6).reshape(-1, 1)
plt.plot(x, Normal().log_prob(x), ls=":", label="Normal")
plt.plot(x, Uniform().log_prob(x), label="Uniform")
plt.plot(x, Beta(peakness=1).log_prob(x), ls="--", label="Beta(1)")
plt.plot(x, Beta(peakness=10).log_prob(x), label="Beta(10)")
plt.plot(x, TruncatedNormal().log_prob(x), label="TruncatedNormal()")
plt.legend();
```
