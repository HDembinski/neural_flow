```python
from neural_flow.distributions import Normal, Uniform, TruncatedNormal, Beta
from matplotlib import pyplot as plt
import numpy as np
import jax
```

```python
x = np.linspace(-0.1, 1.1).reshape(-1, 1)
for dist in (Normal(), Uniform(), Beta(), Beta(peakness=1.0), TruncatedNormal()):
    plt.plot(x, dist.log_prob(x), label=f"{dist!r}")
plt.legend();
```

```python
for dist in (Normal(), Uniform(), Beta(), Beta(peakness=1.0), TruncatedNormal()):
    dist.log_prob(np.zeros((1, 1)))  # initialize
    x = dist.sample(100_000, jax.random.PRNGKey(0))
    plt.figure()
    plt.title(f"{dist}")
    plt.hist(x.reshape(-1))
```

```python
for dist in (Normal(), Uniform(), Beta(), Beta(peakness=1.0), TruncatedNormal()):
    dist.log_prob(np.zeros((1, 2)))  # initialize
    x = dist.sample(10_000, jax.random.PRNGKey(0))
    plt.figure()
    plt.title(f"{dist}")
    plt.scatter(*x.T, marker=".", alpha=0.5, ec="none")
```
